```python
train_ratio = 0.9

import pandas as pd
data = pd.read_csv('input/titanic/train.csv')
train = data.copy()[:int(len(data) * train_ratio)]
val = data.copy()[int(len(data) * train_ratio):]
```

```python
features = ['Pclass', 'Sex', 'Age', 'Age_isnull', 'SibSp', 'Parch', 'Fare', 'C', 'Q', 'S']

def work(df: pd.DataFrame):
    # 将 Embarked 进行 one-hot 编码
    df.Embarked = df.Embarked.dropna()
    df = df.join(pd.get_dummies(df.Embarked))
    df = df.drop('Embarked', axis=1)

    # 将 Sex 替换为 0/1
    df.Sex = df.Sex.replace('male', 0).replace('female', 1)

    q = df.Age.isnull()
    q.name = 'Age_isnull'
    df = df.join(q)
    df.Age = df.Age.fillna(-1)

    df.Fare = df.Fare.fillna(-1)

    return df[features]

train_X = work(train)
train_y = train.Survived
val_X = work(val)
val_y = val.Survived

input_shape = [len(features)]
```
